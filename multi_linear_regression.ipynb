```python
#import liabraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import dataset
ds=pd.read_csv("startup_data.csv")
ds
```

```python
#Divided dataset as Independent and Dependent variables
x=ds[['R&D Spend','Administration','Marketing Spend','State']]
y=ds[['Profit']]
```

```python
# cross check
x
```

```python
#cross check
y
```

```python
ds['State'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
# Initialize the encoder
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')  # safer with unseen categories
# Apply the encoder
encoded_data = ohe.fit_transform(x[['State']])
# Create column names
column_names = ohe.get_feature_names_out(['State'])
# Create DataFrame for encoded columns
encoded_df = pd.DataFrame(encoded_data, columns=column_names, index=x.index)  # maintain index alignment
# Combine with original DataFrame
x = pd.concat([x.drop(columns=['State']), encoded_df], axis=1)
```

```python
#Splitting dataset as Traning and Testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#Fit Model(Multiple Linear Regression MOdel)
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
#predict
data=[[100000,100000,10000,1,0,0]]
columns=x.columns
myinput=pd.DataFrame(columns=columns,data=data)
result=reg.predict(myinput)
result
```

```python
y_pred=reg.predict(x_test)
y_pred
```

```python
#To  ckecj accuracy
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```
