```python
#libraries
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
print("HII")
```

```python
#import 
ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
#splitting datasetinto indepentent and dependent variables
x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
#skfitting random forest Regression Model
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor(n_estimators=10,random_state=0)
reg.fit(x,y)
```

```python
#predict for myinput
reg.predict([[4]])
```

```python
#visualization of random forest regression tree
from sklearn.tree import plot_tree
plot_tree(reg.estimators_[0],filled=True)
plt.show()
```

```python
#predict for user input
level=float(input("Enter The Level:"))
myinput=pd.DataFrame(data=[[level]],columns=["Level"])
result=reg.predict(myinput)
print (f"Predicted salary for level{level}:  ₹{result[0]}")
```
